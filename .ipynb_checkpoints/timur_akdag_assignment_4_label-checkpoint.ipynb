```python
#Examine Labels
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os

label_sheet = 'ko_weekly_labels.csv'
detailed_sheet = 'ko_weekly_return_detailed.csv'
input_dir = r'C:\Users\tiakd\.spyder-py3\Homework 4'
label_file = os.path.join(input_dir, label_sheet)
detailed_file = os.path.join(input_dir,detailed_sheet)


try:   
    with open(label_file) as f:
        ko_dataframe = pd.read_csv(f)
    
    with open(detailed_file) as f:
        ko_all_df = pd.read_csv(f)
    
except Exception as e:
    print(e)
    print('failed to read weekly or detailed file')
```

```python
grouper_year = ko_dataframe.groupby(['Year','Label'])
    
dict_points = dict()
    
for name,group in grouper_year:
    x_axis = np.array(grouper_year.get_group(name)['mean_return'])
    y_axis = np.array(grouper_year.get_group(name)['volatility'])
        
    dict_points[name] = [x_axis,y_axis]
        
  
plt.scatter(dict_points[(2021,'Green')][0],dict_points[(2021,'Green')][1],color = 'green',s=[50 for x in range(len(dict_points[(2021,'Green')][0]))])
plt.scatter(dict_points[(2021,'Red')][0],dict_points[(2021,'Red')][1],color = 'red',s=[50 for x in range(len(dict_points[(2021,'Red')][0]))])
plt.title('2021')
plt.xlabel('mean return')
plt.ylabel('standard deviation')
plt.show()
```

```python
plt.scatter(dict_points[(2022,'Green')][0],dict_points[(2022,'Green')][1],color = 'green',s=[50 for x in range(len(dict_points[(2022,'Green')][0]))])
plt.scatter(dict_points[(2022,'Red')][0],dict_points[(2022,'Red')][1],color = 'red',s=[50 for x in range(len(dict_points[(2022,'Red')][0]))])
plt.title('2022')
plt.xlabel('mean return')
plt.ylabel('standard deviation')
plt.show()
```

```python
week_grp_year = ko_dataframe.groupby('Year')
all_grp = ko_all_df.groupby(['Year','Week_Number'])
    
buget = 100
num_shares = 0
    
first_inv = True

yearly_profit = {}
    
for name,grp in week_grp_year:
    #Loop name = year, grp is the dataframe associated with that year
    for x in range(0,len(grp)):
        #week_series is the current week
        week_series = grp.iloc[x]
            
        #Handle exceptions looking into the future
        if x == len(grp)-1:   
            this_all_week = all_grp.get_group((name,week_series['Week_Number']))
            sell_price = this_all_week.iloc[len(this_all_week)-1]['Adj Close']
            break
        else:
            #Looking into the future is no issue
            nxt_all_first_week = all_grp.get_group((name,week_series['Week_Number']+1)).iloc[0]
            nxt_week = grp.iloc[x+1]   
            #Here we can look into the future and determine what to do
            
        #Current week is label is Red
        if week_series['Label'] == 'Red':   
            #We have not invested yet
            if first_inv:
                continue
                
            if nxt_week['Label'] == 'Green':
                #Invest all money accumulated  
                num_shares = buget / nxt_all_first_week['Open']
                    
                #Go into all_grp for next week's 
            else:
                #This week is red and next week is red no investment
                continue 
        else:
            #Current week is green
            this_all_week = all_grp.get_group((name,week_series['Week_Number']))
                
            if first_inv:
                num_shares = buget / this_all_week.iloc[0]['Open']
                first_inv = False
                
            if nxt_week['Label'] == 'Red':
                #Close out the position
                sell_price = this_all_week.iloc[len(this_all_week)-1]['Adj Close']
                
                #Sell shares
                buget = num_shares * sell_price
                num_shares = 0
            else:
                continue   
    #profit is
    if num_shares > 0:
        year_prft = num_shares * sell_price
    else:
        year_prft = buget
    yearly_profit{name} = year_prft
```

```python
print('Week labeling Q1')
hold_shares = 100 / all_grp.get_group((2021,1)).iloc[0]['Open']       
lst_week =   all_grp.get_group((2022,52))
hold_profit = hold_shares * lst_week.iloc[len(lst_week)-1]['Adj Close']
```

```python
print('Week labeling Q2')
if buget > hold_profit:
    print('Label strategy better')
elif buget == hold_profit:
    print('Buy and hold is as good as label strategy')
else:
    print('Buy and hold results in a larger amount')
```
